```python
import pandas as pd

data = pd.read_csv('user_behavior_dataset.csv')
```

```python
data.head()
```

```python
feature = data.drop(['User ID','Device Model'], axis=1)
feature
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

os_label = labelencoder.fit_transform(feature['Operating System'])

gender_label = labelencoder.fit_transform(feature['Gender'])

feature_date = feature.drop(['Operating System','Gender','User Behavior Class'],axis=1)

feature_date['os'] = os_label
feature_date['gender'] = gender_label

feature_date
```

```python
feature_date.columns
```

```python
target_data = data['User Behavior Class']
target_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature_date, target_data, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# from sklearn.metrics import accuracy_score

accuracy = model.score(X_test,y_test)

print(f"Accuracy of the model: {accuracy:.2f}")
```

```python
input_data = []
para =['What is your minimum App usage time for a day in minutes?\n',
       'What is your screen time for a day in hours?\n',
       'How much battery your using for a day in mAH?\n',
       'How many apps you installed on your mobile?\n',
       'How much data/internet your using for a day in MB?\n',
       'What is your age?\n',
       'What OS your using on your mobile\n0. Andriod\n1. iOS\n',
       'What is your gender\n1. Male\n0. Female\n',]

for x in para:
    input_data.append(int(input(x)))

out = model.predict([input_data])

print("\n\n")

if out[0] < 3:
    print(f"I think you are using your phone less. And my rating of your usage is : {out[0]}")
elif out[0] == 3:
    print(f"I think you are using your phone normally don't use more. And my rating of your usage is : {out[0]}")
elif out[0] > 3:
    print(f"I think you are using too much. you should reduce. And my rating of your usage is : {out[0]}")
else:
    print("Sorry, I can't predict.")

print("\n\n")
```
